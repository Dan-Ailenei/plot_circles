```python
def map_val(x, input_start, input_end, output_start, output_end):
    return output_start + ((output_end - output_start) / (input_end - input_start)) * (x - input_start)
```

```python
def point_on_circle(r, theta):
    return r * np.cos(theta), r * np.sin(theta)
```

```python
DIM = 5
THETA_LENGHT = 2 * np.pi

def make_map(num_circles, fig, ax):
    r = DIM
    ratio = THETA_LENGHT / num_circles
    r_child = r / num_circles
    
    alfa = THETA_LENGHT / num_circles
#     a = r
#     b = a / np.cos(alfa)
#     c = np.sqrt(b ** 2 - a ** 2)
#     r_child = (a + b + c) / (a * b * np.sin(alfa/2))
    
    
    r_child = (r * np.sin(alfa/2)) / (1 + np.sin(alfa/2))
    
    r_imaginary_circle = r - r_child
    
    for i in range(num_circles):
        # draw circle
        theta = map_val(i, 0, num_circles, 0, THETA_LENGHT)
        ax.add_artist(plt.Circle(point_on_circle(r_imaginary_circle, theta), r_child, color='r'))
        
        # draw line
        x1, y1 = point_on_circle(r, theta - ratio / 2)
        x2, y2 = point_on_circle(r, theta - ratio / 2 + np.pi)
        ax.add_line(mlines.Line2D([x1, x2], [y1, y2]))
    fig.show()
```

```python
def init():
    fig, ax = plt.subplots()
    ax.set_xlim((-DIM, DIM))
    ax.set_ylim((-DIM, DIM))
    fig.gca().set_aspect('equal', adjustable='box')

    r = DIM
    t = np.linspace(0, 2 * np.pi)

    x, y = point_on_circle(r, t)
    ax.plot(x, y)
    return fig, ax
```

```python
fig, ax = init()
make_map(10, fig, ax)
```
